```python
from igraph import Graph
from igraph import plot
import numpy as np
```

```python
grafo = Graph(edges = [(0,2), (0,1), (1,4), (1,5), (2,3), (6,7), (3,7), (4,7), (5,6)], directed = True)
grafo.vs["label"] = ["A", "B", "C", "D", "E", "F", "G", "H"]
grafo.es["weight"] = [2,1,2,1,2,1,3,1]
plot(grafo, bbox = (300,300), edge_label = grafo.es["weight"])
```

```python
comunidades = grafo.clusters()
print(comunidades)
```

```python
comunidades.membership
```

```python
c = grafo.community_edge_betweenness()
print(c)
```

```python
c.optimal_count
```

```python
comunidades2 = c.as_clustering()
print(comunidades2)
```

```python
comunidades2.membership
```

```python
plot(grafo, bbox = (300,300), vertex_color = comunidades2.membership)
```

```python
cores = comunidades2.membership
cores
```

```python
cores = np.array(cores)
cores = cores * 100
list(cores)
```
